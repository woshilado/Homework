**Задание 1**  
Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd
```

```python
movies_data = pd.read_csv('movies.csv')
ratings_data = pd.read_csv('ratings.csv')
```

```python
# Решение в одну строку

pd.merge(movies_data, ratings_data, how='left', on='movieId').query('rating == 5.0')['title'].value_counts().head(1)
```

```python
# Более понятное решение (для меня)

titles_and_ratings = pd.merge(movies_data, ratings_data, how='left', on='movieId') # объединяем таблицы по movieId
titles_and_ratings_5 = titles_and_ratings.query('rating == 5.0') # создаем список только с пятерками
titles_and_ratings_5['title'].value_counts().head(1) # считаем количество упоминаний названия фильма, выводим первый
```

**Задание 2**  
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
data = pd.read_csv('power.csv')
data.head()
```

```python
baltics = data[ (data['country']=='Lithuania') | (data['country']=='Latvia') | (data['country']=='Estonia') ]
baltics.head()
```

```python
baltics_cat = baltics[ (baltics['category']==4) | (baltics['category']==12) | (baltics['category']==21) ]
baltics_cat.head()
```

```python
baltics_cat_year = baltics_cat[ (2005 <= baltics_cat['year']) & (baltics_cat['year'] <= 2010) ]
baltics_cat_year.head()

# почему не работает это?
# baltics_cat_year = baltics_cat[ 2005 <= baltics_cat['year'] <= 2010 ]
```

```python
baltics_cat_year_pos = baltics_cat_year[ baltics_cat_year['quantity']>= 0 ]
baltics_cat_year_pos.head()
```

```python
baltics_cat_year_pos['quantity'].sum() # or baltics_cat_year_pos.quantity.sum()
```
