**Наблюдения и предложения:**
0. Поскольку предстоит определить ответ водителя (да/нет), наиболее подходящей моделью будет RandomForest. Из этого мы и будем исходить при дальнейшем инжиниринге признаков.
1. Номера, присвоенные предложениям и заказам, можно смело удалить. Они не несут никакой смысловой нагрузки.
2. Номер водителя гораздо более важный показатель, поскольку определенные водители имеют собственные предпочтения, соответственно, отказываются от заказов чаще или реже.
3. Широту и долготу точек нахождения водителя и заказа можно удалить (4 признака), поскольку имеется более информативный признак, в полной мере их заменяющий - расстояние от водителя до места заказа. Если бы его не было, то его обязательно стоило бы создать с помощью координат точек и теоремы Пифагора. Но вместе с этим есть время поездки до места заказа, которое сильно коррелирует с расстоянием. Одно из них стоит удалить. Попрощаемся с расстоянием, поскольку время - золото.
4. Во времени поездки есть странные нулевые и отрицательные значения. Можно было бы предположить, что водитель уже находился на месте на момент заказа (аэропорты, отели, торговые и бизнес-центры), но значения широт и долгот говорят об обратном. От них придется избавиться.
5. Оставшиеся признаки переводим в числовые значения через LabelEncoder.

Удаляем лишние признаки

```python
data.drop(['offer_gk', 'order_gk', 'driver_latitude', 'driver_longitude',
           'origin_order_latitude', 'origin_order_longitude', 'distance_km'], axis=1, inplace=True)
data.head()
```

Удалим строки, содержащие отрицательные или нулевые значения в столбце duration_min

```python
data = data.loc[~(data.duration_min == -1)|(data.duration_min == 0)]
data.reset_index(drop=True, inplace=True)
data.head()
```

Преобразуем столбец duration_min в более удобный вид, разбив его на пятиминутки, т.е. мы будем иметь столбец с количеством пятиминуток, которое предстояло проехать водителю до клиента. Такой маневр уменьшает количество категорий в признаке до 73.

```python
data['duration'] = data.duration_min.apply(lambda x: int(x // 5))
data.drop(['duration_min'], axis=1, inplace=True)
data.head()
```

C помощью LabelEncoder преобразуем категориальные значения в числа

```python
from sklearn.preprocessing import LabelEncoder

label_enc = LabelEncoder()
data['labeled_offer_class'] = label_enc.fit_transform(data['offer_class_group'])
data['labeled_ride_type'] = label_enc.fit_transform(data['ride_type_desc'])
data.drop(['offer_class_group', 'ride_type_desc'], axis=1, inplace=True)
data.head()
```

Разделим driver_response на прием заказа и отказ (positive_response и negative_response)

```python
data['positive_response'] = data['driver_response'].apply(lambda x: 1 if x == 1 else 0)
data['negative_response'] = data['driver_response'].apply(lambda x: 1 if x == 0 else 0)
data.drop(['labeled_offer_class'], axis=1, inplace=True)
data.head()
```

Отделим зависимую переменную от независимых (регрессоров)

```python
input_x = data.iloc[:, :6]
output_y = data.iloc[:, 6:]
```

Создадим одномерный массив для зависимой переменной

```python
label_enc = LabelEncoder()
targets = (output_y.iloc[:,:]==1).idxmax(1)
Y = label_enc.fit_transform(targets)
print(targets.value_counts())

# P.S. в принципе, можно было и не разбивать признак driver_response на два, а сразу поместить его в Y
```

Разделим данные на тренировочные и тестовые

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(input_x, Y, test_size=0.2)
```

Создаем модель случайного леса и выбираем 100 деревьев в качестве параметра. Затем обучаем модель

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100)
model.fit(X_train, y_train)
```

Делаем classification report и оцениваем качество модели с помощью коэффициента детерминации

```python
from sklearn.metrics import classification_report

y_test_predicted = model.predict(X_test)
print(classification_report(y_test, y_test_predicted))

model.score(X_test, y_test)
```
