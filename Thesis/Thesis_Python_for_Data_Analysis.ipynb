```python
# Мода (третий вариант)

# В качестве примера рассмотрим колонку satisfaction_level
# Создадим пустой словарь, в котором будем суммировать все появления определенных значений

sat_lev_counts = {}
for x in df.satisfaction_level:
    if x not in sat_lev_counts:
        sat_lev_counts[x] = 1
    else:
        sat_lev_counts[x] += 1
        
# Найдем в словаре максимальное количество повторений с помощью алгоритма нахождения максимума

max_x = 0
for k, v in sat_lev_counts.items():
    if max_x < v:
        max_x = v
        mode_sat_lev = k
print('Значение моды:', mode_sat_lev, '\nКоличество встречаемости:', max_x)
```

```python
# Среднее квадратическое отклонение (второй вариант)

# В качестве примера рассмотрим колонку satisfaction_level

def stdev(nums):
    diffs = 0
    avg = sum(nums)/len(nums) # считаем среднее значение
    for n in nums:        
        diffs += (n - avg)**(2) # считаем сумму квадратичных отклонений
    return (diffs/(len(nums)-1))**(0.5) # считаем корень среднеквадратичного значения

stdev(df.satisfaction_level)
```

**5. С использованием t-test статистики проверить взаимосвязи между несколькими переменными**

```python
# Для наглядности проверим статистическое отличие между переменными из предыдущего задания
# Для одновыборочного t-критерия: stats.ttest_1samp(), для двухвыборочного: stats.ttest_ind()
# Поскольку мы рассматриваем разные переменные (а не выборку и ген. совокупность), воспользуемся двухвыборочным t-критерием

from scipy import stats

print(f't-value для среднего количества часов и номера проекта: {stats.ttest_ind(df.average_montly_hours, df.number_project)[0]}')
print(f'p-value для среднего количества часов и номера проекта: {stats.ttest_ind(df.average_montly_hours, df.number_project)[1]}\n')
print(f't-value для среднего количества часов и последней оценки: {stats.ttest_ind(df.average_montly_hours, df.last_evaluation)[0]}')
print(f'p-value для среднего количества часов и последней оценки: {stats.ttest_ind(df.average_montly_hours, df.last_evaluation)[1]}\n')
print(f't-value для последней оценки и стажа в компании: {stats.ttest_ind(df.last_evaluation, df.time_spend_company)[0]}')
print(f'p-value для последней оценки и стажа в компании: {stats.ttest_ind(df.last_evaluation, df.time_spend_company)[1]}\n')
print(f'Количество рассматриваемого персонала: {len(df.average_montly_hours)}')
```

**10. Проверить несколько гипотез для данных, чтобы понять почему сотрудники покидают компанию**

```python
# Корреляция Пирсона

# H0: зависимость между данными отсутствует
# H1: зависимость между данными существует

from scipy.stats import pearsonr

stat, pvalue = pearsonr(df.left, df.time_spend_company)
print('stat=%.3f, p=%.3f' % (stat, pvalue))
if pvalue > 0.05:
    print('Принимаем Н0. Зависимость между уходом из компании и временем, проведенном в ней, скорее отсутствует')
else:
    print('Отвергаем Н0. Зависимость между уходом из компании и временем, проведенном в ней, скорее существует')
```

```python
# Ранговая корреляция Спирмена

# H0: зависимость между данными отсутствует
# H1: зависимость между данными существует

from scipy.stats import spearmanr

stat, pvalue = spearmanr(df.left, df.number_project)
print('stat=%.3f, p=%.3f' % (stat, pvalue))
if pvalue > 0.05:
    print('Принимаем Н0. Зависимость между уходом из компании и номером проекта скорее отсутствует')
else:
    print('Отвергаем Н0. Зависимость между уходом из компании и номером проекта скорее существует')
```

```python
# Ранговая корреляция Кендалла

# H0: зависимость между данными отсутствует
# H1: зависимость между данными существует

from scipy.stats import kendalltau

stat, pvalue = kendalltau(df.left, df.last_evaluation)
print('stat=%.3f, p=%.3f' % (stat, pvalue))
if pvalue > 0.05:
    print('Принимаем Н0. Зависимость между уходом из компании и последней оценкой скорее отсутствует')
else:
    print('Отвергаем Н0. Зависимость между уходом из компании и последней оценкой скорее существует')
```
