### Задание 1

Получить интуицию, на что влияют параметры распределений. Построить по 5 функций плотностей с разными параметрами для каждого распределения: нормальное, экспоненциальное, Стьюдента на одном графике.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
```

Плотность вероятности (probability density function - PDF)

**Нормальное распределение**

```python
# зададим среднее значение "мю"
mu1 = 2.0
mu2 = 2.0
mu3 = 3.0
mu4 = 4.0
mu5 = 4.0

# зададим среднеквадратичное отклонение "сигма"
sigma1 = 0.3
sigma2 = 0.4
sigma3 = 0.5
sigma4 = 0.6
sigma5 = 1

# зададим нормально распределенную случайную величину
norm_rv1 = sts.norm(loc=mu1, scale=sigma1)
norm_rv2 = sts.norm(loc=mu2, scale=sigma2)
norm_rv3 = sts.norm(loc=mu3, scale=sigma3)
norm_rv4 = sts.norm(loc=mu4, scale=sigma4)
norm_rv5 = sts.norm(loc=mu5, scale=sigma5)

# сгенерируем 10 значений
# norm_rv1.rvs(size=10)
```

```python
x = np.linspace(0,7,100)

pdf1 = norm_rv1.pdf(x)
pdf2 = norm_rv2.pdf(x)
pdf3 = norm_rv3.pdf(x)
pdf4 = norm_rv4.pdf(x)
pdf5 = norm_rv5.pdf(x)

plt.plot(x, pdf1)
plt.plot(x, pdf2)
plt.plot(x, pdf3)
plt.plot(x, pdf4)
plt.plot(x, pdf5)

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
```

**Экспоненциальное распредление**

```python
# зададим единственный параметр данного распределения "лямбда"
lam1 = 0.1
lam2 = 0.5
lam3 = 1
lam4 = 1.5
lam5 = 2

# зададим экспоненциально распределенную случайную величину
expon_rv1 = sts.expon(lam1)
expon_rv2 = sts.expon(lam2)
expon_rv3 = sts.expon(lam3)
expon_rv4 = sts.expon(lam4)
expon_rv5 = sts.expon(lam5)

# сгенерируем 10 значений
# expon_rv1.rvs(10)
```

```python
x = np.linspace(0,6,1000)

pdf1 = expon_rv1.pdf(x)
pdf2 = expon_rv2.pdf(x)
pdf3 = expon_rv3.pdf(x)
pdf4 = expon_rv4.pdf(x)
pdf5 = expon_rv5.pdf(x)

plt.plot(x, pdf1)
plt.plot(x, pdf2)
plt.plot(x, pdf3)
plt.plot(x, pdf4)
plt.plot(x, pdf5)

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
```

**Распределение Стьюдента**

```python
# зададим число степеней свободы n
n1 = 1
n2 = 2
n3 = 3
n4 = 10
n5 = 100

# зададим случайную величину c распределением Стьюдента
t_rv1 = sts.t(n1)
t_rv2 = sts.t(n2)
t_rv3 = sts.t(n3)
t_rv4 = sts.t(n4)
t_rv5 = sts.t(n5)

# сгенерируем 10 значений
# t_rv1.rvs(10)
```

```python
x = np.linspace(-5,5,1000)

pdf1 = t_rv1.pdf(x)
pdf2 = t_rv2.pdf(x)
pdf3 = t_rv3.pdf(x)
pdf4 = t_rv4.pdf(x)
pdf5 = t_rv5.pdf(x)

plt.plot(x, pdf1)
plt.plot(x, pdf2)
plt.plot(x, pdf3)
plt.plot(x, pdf4)
# plt.plot(x, pdf5)

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
```

Посмотрим на вышеописанные распределения на одном графике

```python
# Нормальное распределение
mu = 2.0
sigma = 0.5
norm_rv = sts.norm(loc=mu, scale=sigma)

# Экспоненциальное распредление
lam = 1
expon_rv = sts.expon(lam)

# Распределение Стьюдента
n = 3
t_rv = sts.t(n)
```

```python
x = np.linspace(-5,7,1000)
pdf_norm = norm_rv.pdf(x)
pdf_expon = expon_rv.pdf(x)
pdf_t = t_rv.pdf(x)

plt.plot(x, pdf_norm)
plt.plot(x, pdf_expon)
plt.plot(x, pdf_t)

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
```
